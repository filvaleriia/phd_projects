# Funktion for calculate metrics

```python
def calculate_metrics(generated_set, test_set, name, var):
    results = pd.DataFrame(columns = ['var','uniq_scaffolds','set_size','tpra','tpra','tprar','tprar',\
                                 'sescy','sescr','sescry','asescr'])
    mt = metrics_update.Metrics(var)        
    name_generated_set = generated_set
    name_test_set = test_set
    mt.load(name_generated_set, name_test_set)
    #mt.calculate_metrics()
    result = mt.calculate_metrics_with_return()
    results.loc[len(results)] = result
    results.insert(loc=0, column='name', value=[name])

    df_generated, input1 = mt.unique()

    return results,df_generated,input1
```
